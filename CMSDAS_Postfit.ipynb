```python
MCsys16low = []
MCsys16up = []
#open the file and grab the histograms
fn = '/eos/user/c/cfreer/Moriond_2020/CMSSW_10_2_13/src/MonoZFinalFit/fitDiagnosticsMonoZ_Yields_DMSimp.root'
f2 = ROOT.TFile.Open("/eos/user/c/cfreer/Moriond_2020/CMSSW_10_2_13/src/MonoZFinalFit/fitDiagnosticsMonoZ_Yields_DMSimp.root")
f = uproot.open(fn)
histograms = f.allitems( filterclass=lambda cls: issubclass(cls, uproot_methods.classes.TH1.Methods))

for name, roothist in histograms:
    name = name.decode("utf-8")
    name = name.replace(";1", "")
    era = "2016"
    channel = "cat4L"
    #fit_type = "prefit"
    fit_type = "fit_b"
    if  "shapes_"+fit_type+"/"+channel+"2016/ZZ" in name:
            ZZ_freq = np.abs(roothist.numpy())[0]
    if  "shapes_"+fit_type+"/"+channel+"2016/WZ" in name:
            WZ_freq = np.abs(roothist.numpy())[0]
    if  "shapes_"+fit_type+"/"+channel+"2016/WW" in name:
            WW_freq = np.abs(roothist.numpy())[0]
    if  "shapes_"+fit_type+"/"+channel+"2016/TOP" in name:
            TOP_freq = np.abs(roothist.numpy())[0]
    if  "shapes_"+fit_type+"/"+channel+"2016/DY" in name:
            DY_freq = np.abs(roothist.numpy())[0]
    if  "shapes_"+fit_type+"/"+channel+"2016/VVV" in name:
            VVV_freq = np.abs(roothist.numpy())[0]
            
    if  "shapes_"+fit_type+"/"+channel+"2016/total_background" in name:
            bg_freq = np.abs(roothist.numpy())[0]
            histog = f2.Get("shapes_"+fit_type+"/"+channel+"2016/total_background")
            for i in range(int(histog.GetEntries())):
                MCsys16low.append(histog.GetBinErrorLow(i))
                MCsys16up.append(histog.GetBinErrorUp(i))

    if  "shapes_"+fit_type+"/"+channel+"2016/data" in name:
            data_freq16 = np.abs(roothist.numpy())[0]

#error with 0 entries so just set to 0
WW_freq = ZZ_freq * 0

MCsys16low = np.array(MCsys16low)
MCsyslow = MCsys16low[:13]
MCsys16up = np.array(MCsys16up)
MCsysup = MCsys16up[:13]

#Grab the TGraph for data            
f2 = ROOT.TFile.Open("/eos/user/c/cfreer/Moriond_2020/CMSSW_10_2_13/src/MonoZFinalFit/fitDiagnosticsMonoZ_Yields_DMSimp.root")
data_hist16 = f2.Get("shapes_"+fit_type+"/"+channel+"2016/data")
npoints = data_hist16.GetN()
data_set = []
Yhigh = []
Ylow = []
values16 = data_hist16.GetY()
alpha = 1. - 0.6827
for i in range(npoints):
    value = values16[i]
    data_set.append(value)
    #Calculate Poisson errors
    if value == 0 : L = 0
    if value != 0 : L = ROOT.Math.gamma_quantile(alpha/2,value,1.)
    U = ROOT.Math.gamma_quantile_c(alpha/2,value+1,1)
    Ylow.append(value-L)
    Yhigh.append(U-value)
    #print(L,U)
```

```python
ZZ_freq = ZZ_freq[:12]
WZ_freq = WZ_freq[:12]
WW_freq = WW_freq[:12]
DY_freq = DY_freq[:12]
TOP_freq = TOP_freq[:12]
VVV_freq = VVV_freq[:12]
bg_freq = bg_freq[:12]
MCsyslow = MCsyslow[:13]
MCsyslow = MCsyslow[1:]
MCsysup = MCsysup[:13]
MCsysup = MCsysup[1:]

data_set = np.array(data_set)
Yhigh = np.array(Yhigh)
Ylow = np.array(Ylow)
data_set = data_set[:12]
Yhigh = Yhigh[:12]
Ylow = Ylow[:12]
```

```python
#Set up the bins
bins = np.array([100,125,150,175,200,250,300,350,400,500,600,1000,1000])
binc = np.array([ 0.5*(bins[i]+bins[i+1])for i in range(bins.shape[0]-1)])
xerr = np.diff(bins)*0.5

print (np.size(bins))
print (np.size(bg_freq))

#Set up variables for the stacked histogram
bg_freq = VVV_freq + ZZ_freq + WW_freq + WZ_freq + TOP_freq + DY_freq
freqs = [VVV_freq, ZZ_freq, WW_freq, WZ_freq, TOP_freq, DY_freq]
binsc = [binc, binc, binc, binc, binc, binc]
proc_name = ["VVV", "ZZ", "WW", "WZ", "TOP", "DY"]
colors = ["maroon", "gold", mcd.XKCD_COLORS["xkcd:green"].upper(), mcd.XKCD_COLORS["xkcd:azure"].upper(), mcd.XKCD_COLORS["xkcd:darkblue"].upper(), mcd.XKCD_COLORS["xkcd:pink"].upper()]

#Set up variables for the stacked histogram
plt.figure(figsize=(8,10))
plt.gcf().subplots_adjust(bottom=0.15, left=0.17)
ax1 = plt.subplot2grid((4,1), (0,0),rowspan=2)

#the histograms (SM,Data)
ax1.hist(binsc,bins=bins,weights=freqs,stacked=True, color=colors, label=proc_name)
ax1.errorbar(binc,data_set, xerr=xerr, yerr = [Ylow,Yhigh],color="black", fmt='o')
ax1.bar(binc,MCsyslow+MCsysup,bottom=bg_freq-MCsyslow,color='gray', hatch='',width=xerr*2,alpha=0.3)

#Labels and limits
ax1.set_yscale("log")
ax1.set_ylim(0.001,5000)
ax1.legend(loc="upper right", fontsize=14)
ax1.set_ylabel(r"$events$")
cms = ax1.text(
        0.5, 5500., u"CMS $\it{preliminary}$",
        fontsize=16, fontweight='bold',
    )
lumi = ax1.text(
        800, 5500., r"%.1f fb$^{-1}$ (13 TeV)" % 35.9,
        fontsize=14, 
    )


#Adding in the ratio plot
ax2 = plt.subplot2grid((4,1), (2,0),sharex=ax1)
plt.setp(ax1.get_xticklabels(), visible=False)
ax2.errorbar(binc,data_set/bg_freq, xerr=xerr,yerr=[data_set/bg_freq - (data_set-Ylow)/bg_freq,(data_set+Yhigh)/bg_freq - data_set/bg_freq], color="black", fmt='o')

ax2.bar(binc,(MCsyslow+MCsysup)/bg_freq,bottom=1-MCsyslow/bg_freq,color='gray',width=xerr*2,alpha=0.3,hatch='')
ax2.axhline(1, ls="--")
ax2.set_ylim(0,4)
ax2.set_xlabel("$p_{T}^{miss}$ (GeV)",ha="right")
#ax2.set_xlabel("$M_{T}$ (GeV)",ha="right")
ax2.set_ylabel(r"Data/$\Sigma$MC")
```
